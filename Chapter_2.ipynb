```python
boston_path = '/content/drive/MyDrive/Data Science/data/Boston.csv'
boston_df = pd.read_csv(boston_path)
boston_df.tail()
```

## b)

```python
boston_df.shape
```

The dataset has 506 rows and 14 columns. The first column seems to be an identifier. Each row is an observation of the following variables:

1. CRIM - per capita crime rate by town
2. ZN - proportion of residential land zoned for lots over 25,000 sq.ft.
3. INDUS - proportion of non-retail business acres per town.
4. CHAS - Charles River dummy variable (1 if tract bounds river; 0 otherwise)
5. NOX - nitric oxides concentration (parts per 10 million)
6. RM - average number of rooms per dwelling
7. AGE - proportion of owner-occupied units built prior to 1940
8. DIS - weighted distances to five Boston employment centres
9. RAD - index of accessibility to radial highways
10. TAX - full-value property-tax rate per \$10,000
11. PTRATIO - pupil-teacher ratio by town
13. LSTAT - % lower status of the population
14. MEDV - Median value of owner-occupied homes in \$1000's


## c)

```python
boston_df = boston_df.drop('Unnamed: 0', axis=1)
boston_df.tail()
```

```python
sns.pairplot(data=boston_df.iloc[:,:6])
plt.show()
```

None of the first 6 columns seems to be related with each other.

## d)

```python
fig, ax = plt.subplots(12, 1, figsize = (8,32), dpi=150)

for i,col in enumerate(list(boston_df.columns)[1:]):
  sns.scatterplot(data=boston_df, x=col, y='crim', ax=ax[i])
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=boston_df, x='chas', y='crim')
plt.show()
```

The age and dis variables seem to have an exponetial relation with crim, whereas lstat and medv show a more linear relation.

## e)

```python
fig, ax = plt.subplots(3, 1, figsize=(4,12), dpi=150)
sns.boxplot(data=boston_df, y='crim', ax=ax[0])
sns.boxplot(data=boston_df, y='tax', ax=ax[1])
sns.boxplot(data=boston_df, y='ptratio', ax=ax[2])
plt.show()
```

There is a lot of outliers in the crim variable. That means that many towns have a high crime rate. The tax and pupil-teacher ratios do not show this behavior.
